## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
client_data["subtotal"] = client_data["unit_price"] * client_data["qty"]
client_data.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
# client_data["shipping price"] = lambda unit_weight, qty: 7 * unit_weight if unit_weight * qty > 50 else 10  )
# client_data["shipping price"] = lambda x: 7 * (client_data["qty"] * client_data["unit_weight"]) if client_data["unit_weight"] * client_data["qty"] > 50 else 10 * (client_data["qty"] * client_data["unit_weight"])
# cal_shipping_cost = lambda unit_weight, qty: 7 * (qty * unit_weight) if (unit_weight * qty) > 50 else 10 * (qty * unit_weight)
# client_data['shipping_price'] = cal_shipping_cost(client_data['unit_weight'], client_data['qty'])
cal_shipping_cost = lambda x: 7 * (x['qty'] * x['unit_weight']) if (x['unit_weight'] * x['qty']) > 50 else 10 * (x['qty'] * x['unit_weight'])
client_data['shipping price'] = client_data.apply(cal_shipping_cost, axis=1)

client_data.head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
sales_tax = 1.0925
total_price = (client_data["subtotal"] + client_data["shipping price"]) * float(sales_tax)
client_data["total price"] = round(total_price, 2)
client_data.head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
client_data["new total cost"] = (client_data["unit_cost"] * client_data["qty"]) + client_data["shipping price"]
client_data.head()
```

```python
# Create a column for the profit of
# each line using line cost and line price
client_data["profit"] = client_data["total price"] - client_data["new total cost"]
client_data
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
top_clients = client_data.groupby("client_id")["total price"].sum()
top5_clients = top_clients[top_5_clients_df]
top5_clients
```

```python
# Create a summary DataFrame showing the totals for the
# working with Nic on this code
summery_of_data = client_data[client_data["client_id"].isin(top_5_clients_df)]
# for the top 5 clients with the following information:
summery_of_data_df = summery_of_data.groupby("client_id").agg({"qty": "sum", 
                                                               "shipping price": "sum", 
                                                               "total price": "sum", 
                                                               "new total cost": "sum", 
                                                               "profit": "sum", }).reset_index(drop=False)
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
summery_of_data_df.sort_values(by="profit", ascending=False)
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
summery_of_data_df = summery_of_data_df.rename(columns={"client_id": "Client ID", 
                                                        "qty": "Units", 
                                                        "shipping price": "Shipping", 
                                                        "total price": "Total Revenue", 
                                                        "new total cost": "Total Cost", 
                                                        "profit": "Total Profit", 
                                                       })
# Currency should be in millions of dollars. 
# Creating a function to set to millions.
def set_to_million(x):
    return x/1000000

# dividing results by 1,000,000 to get correct units
summery_of_data_df["Shipping"] = summery_of_data_df["Shipping"].apply(set_to_million)
summery_of_data_df["Total Revenue"] = summery_of_data_df["Total Revenue"].apply(set_to_million)
summery_of_data_df["Total Profit"] = summery_of_data_df["Total Profit"].apply(set_to_million)
summery_of_data_df["Total Cost"] = summery_of_data_df["Total Cost"].apply(set_to_million)

# Creating a function for format "$" & "M"
def format_millions(x):
    return "${:.2f}M".format(x)

# Adding format to column results
summery_of_data_df["Shipping"] = summery_of_data_df["Shipping"].apply(format_millions)
summery_of_data_df["Total Revenue"] = summery_of_data_df["Total Revenue"].apply(format_millions)
summery_of_data_df["Total Profit"] = summery_of_data_df["Total Profit"].apply(format_millions)
summery_of_data_df["Total Cost"] = summery_of_data_df["Total Cost"].apply(format_millions)

summery_of_data_df
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
summery_of_data_df.sort_values(by="Total Profit", ascending=False)
```
